```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
dataframe = pd.read_csv('C:\\Users\MRINALINI\Downloads\like-unlike\like-unlike\dataset.csv') 
dataframe
```

```python
dataframe.info
```

```python
df= pd.DataFrame(dataframe, columns=['Architecture','Aspect Ratio','Audio Jack','Autofocus','Bezel-less display','Bluetooth','Video Recording','VoLTE','Weight','Wi-Fi Features','Width','Rating'])
df
```

```python
df ['Rating'] = np.where(df['Rating'] > 4.0, 'like', 'dislike') 
df
```

```python
le = LabelEncoder()
df['Rating'] =le.fit_transform(df['Rating'])
df
```

```python
df.dropna(axis=0, how='any',inplace=True)
df
```

```python
df['Wi-Fi Features'] =le.fit_transform(df['Wi-Fi Features'])
df['Width'] =le.fit_transform(df['Width'])
df['Weight'] =le.fit_transform(df['Weight'])
df['VoLTE'] =le.fit_transform(df['VoLTE'])
df['Video Recording'] =le.fit_transform(df['Video Recording'])
df['Architecture'] =le.fit_transform(df['Architecture'])
df['Autofocus'] =le.fit_transform(df['Autofocus'])
df['Bezel-less display'] =le.fit_transform(df['Bezel-less display'])
df['Audio Jack'] =le.fit_transform(df['Audio Jack'])
df['Bluetooth'] =le.fit_transform(df['Bluetooth'])
df['Aspect Ratio'] =le.fit_transform(df['Aspect Ratio'])
df
```

```python
x=df.iloc[:,0:12]
y=df.iloc[:,-1]
Xtrain, Xtest,Ytrain,Ytest= train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(Xtrain,Ytrain)
```

```python
prediction = rf.predict(Xtest)
prediction
```

```python
accuracy_score(rf.predict(Xtest),Ytest)
```

```python
confusion_matrix(Ytest, prediction)
```
